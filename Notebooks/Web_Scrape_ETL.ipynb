# Element extraction

```python
# Clean user info
def clean_user_date_info(text):
    # Use regex to find patterns like 'Name from Location, Date'
    match = re.search(r"(.+), (\d{2}/\d{2}/\d{4})", text)
    if match:
        user_info = match.group(1)  # 'Name from Location'
        date = match.group(2)  # 'Date'
        # The remaining part is the car model, which we assume starts after the date
        car_model = text.split(date)[-1]
        return user_info, date, car_model
    else:
        # If the pattern is not found, return the original text and None values
        return text, None, None

# Extract rating breakdown info
def extract_ratings_from_text(text, attributes):
    ratings = {attr: 0 for attr in attributes}  # Initialize dictionary with attributes
    for attr in attributes:
        # Use a raw string for the regular expression pattern
        match = re.search(rf"{attr}(\d+) out of 5 stars", text, re.IGNORECASE)
        if match:
            # Convert the captured rating to an integer and assign it to the appropriate attribute
            ratings[attr] = int(match.group(1))
    return ratings

def clean_review_text(text):
    # Check if '.Safety' exists in the text
    if '.Safety' in text:
        # Remove everything after '...another Tesla.' since that's where the ratings start
        review_text, _ = text.split('.Safety', 1)
    else:
        # If '.Safety' is not found in the text, return the text as is
        review_text = text
    review_text = review_text.strip('.')  # Remove any trailing dots
    return review_text

# Collection loop
def reviews_extraction(source, review_containers):
    shared_attributes = ['comfort', 'interior', 'performance', 'value', 'reliability']
    reviews_data = []

    if source == 'Cars.com':
        
        for review_container in review_containers:
            # Extract the star rating
            rating = review_container.find('span', class_='sds-rating__count').get_text(strip=True)
                
            # Extract the review title
            title = review_container.find('h3', class_='spark-heading-6').get_text(strip=True)
                
            # Extract the review date and user's name
            date_posted = review_container.find('div', class_='review-byline').find('div').get_text(strip=True)
            user_name = review_container.find('div', class_='review-byline').find_all('div')[1].get_text(strip=True)
                
            # Extract the review text
            review_text = review_container.find('p', class_='review-body').get_text(strip=True)
                
            # Extract the rating breakdown
            rating_breakdown = {}
            breakdown_list = review_container.find('ul', class_='sds-definition-list')
            if breakdown_list:
                for item in breakdown_list.find_all('li', recursive=False):
                    category_text = item.find('span', class_='sds-definition-list__display-name').get_text(strip=True)
                    category = category_text.lower().replace(" ", "_")  # Replace spaces with underscores for consistency
                    if category in shared_attributes:
                        value = item.find('span', class_='sds-definition-list__value').get_text(strip=True)
                        rating_breakdown[category] = value
                
            # Compile the extracted data into a dictionary
            review_data = {
                'overall_rating': rating,
                'title': title,
                'date_posted': date_posted,
                'user_name': user_name,
                'review_text': review_text,
                **rating_breakdown  # Merge the rating breakdown into the review_data
            }
        
            # Append the dictionary to the reviews list
            reviews_data.append(review_data)

    else:
         
        for review_container in review_containers:
            # Extract overall rating
            overall_rating_element = review_container.find('span', class_='sr-only')
            overall_rating_text = overall_rating_element.get_text(strip=True)
            overall_rating = overall_rating_text.split()[0]
    
            # Extract review metadata
            review_title = review_container.find('h2', class_='heading-5').get_text(strip=True)
            user_date_info = review_container.find('div', class_='small text-gray-dark mb-2').get_text(strip=True)
    
            # Clean user date info
            user_info, date, car_model = clean_user_date_info(user_date_info)

            # Extract review text
            review_text = review_container.find('div', class_='truncated-text').get_text(strip=True)
    
            # Extract rating breakdown info
            rating_breakdown = extract_ratings_from_text(review_text, shared_attributes)
    
            # Clean the review text to remove the breakdown
            review_text = clean_review_text(review_text)
    
            # Compile the extracted data into a dictionary
            review_data = {
                'overall_rating': overall_rating,
                'title': review_title,
                'date_posted': date,  
                'user_name': user_info,  
                'review_text': review_text,
                **rating_breakdown 
            }
    
            # Append the dictionary to the reviews list
            reviews_data.append(review_data)
            
    return reviews_data
```
